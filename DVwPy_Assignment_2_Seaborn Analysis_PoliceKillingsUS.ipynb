## 2: Plot 12 most dangerous cities

```python
kill.head()
```

```python
city_freq = kill.city.value_counts()[:12]
city_freq
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(x=city_freq.index, y=city_freq.values)
ax.set_title("Most dangerous cities", color="blue", fontsize=16)
ax.set_xticklabels(city_freq.index, rotation=45);
```

```python
# with seaborn countplot

fig, ax =plt.subplots(figsize=(8,5))
ax = sns.countplot(x=kill.city, order= city_freq.index) 
ax.set_title("Most dangerous cities", color="blue", fontsize=16)
ax.set_xticklabels(city_freq.index, rotation=45);
```

## 3: Plot 20 most dangerous states

```python
state_freq = kill.state.value_counts()[:20]
state_freq
```

```python
# with seaborn barplot
fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(data=kill, x=state_freq.index, y=state_freq.values)
ax.set_xticklabels(state_freq.index , rotation=45)
ax.set_title("Most Dangerous States", fontsize=15, color= "blue");
```

```python
# with seaborn countplot

fig, ax =plt.subplots(figsize=(8,5))
ax = sns.barplot(x=state_freq.index,y=state_freq.values)
ax.set_title("Most dangerous states",color="b",fontsize=16)
ax.set_xticklabels(state_freq.index,rotation=45);
```
